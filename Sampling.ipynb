```scala
import org.apache.spark.sql.types._
import org.apache.spark.sql.ForeachWriter
import org.apache.spark.sql.catalyst.ScalaReflection
import org.apache.spark.sql.Row
import org.apache.spark.sql.functions._
import org.apache.spark.sql.SQLContext

val sqlContext = new SQLContext(sc)
import sqlContext.implicits._
```

```scala
case class Entry(distType : String, num : Float)

var schema = ScalaReflection.schemaFor[Entry].dataType.asInstanceOf[StructType]
```

```scala
var streamUniform = spark.readStream.
    schema(schema).
    csv("dist/")
```

```scala
streamUniform.reservoir(100).groupBy("distType").agg(avg($"num")).writeStream.format("console").outputMode("update").start
```
